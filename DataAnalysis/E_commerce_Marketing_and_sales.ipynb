#**Adding invoice_value column in online_Sales_df**

```python
# 1. Extract month from Transaction_Date in online_Sales_df for merging
online_Sales_df['Transaction_Month'] = online_Sales_df['Transaction_Date'].dt.month

# 2. Extract month from Month in discount_coupon_df for merging
discount_coupon_df['Discount_Month'] = discount_coupon_df['Month'].dt.month

# 3. Merge online_Sales_df with discount_coupon_df
#    Use a left merge to keep all sales records
merged_sales_discount_df = pd.merge(
    online_Sales_df,
    discount_coupon_df[['Discount_Month', 'Product_Category', 'Discount_pct']],
    how='left',
    left_on=['Transaction_Month', 'Product_Category'],
    right_on=['Discount_Month', 'Product_Category'],
    suffixes=('_sales', '_discount')
)

# 4. Initialize a column for effective discount percentage
merged_sales_discount_df['Effective_Discount_pct'] = 0.0

# 5. Apply Discount_pct only if Coupon_Status is 'Used'
#    If a coupon was used but no matching discount was found (NaN from merge), it remains 0.
merged_sales_discount_df.loc[
    merged_sales_discount_df['Coupon_Status'] == 'Used',
    'Effective_Discount_pct'
] = merged_sales_discount_df['Discount_pct']

# Fill any remaining NaN in Effective_Discount_pct (e.g., if 'Used' coupon had no match)
merged_sales_discount_df['Effective_Discount_pct'] = merged_sales_discount_df['Effective_Discount_pct'].fillna(0.0)


# 6. Merge with tax_amount_df to get GST
final_df = pd.merge(
    merged_sales_discount_df,
    tax_amount_df[['Product_Category', 'GST']],
    how='left',
    on='Product_Category'
)

# Fill any potential NaN in GST if a product category didn't have a tax entry
final_df['GST'] = final_df['GST'].fillna(0.0)

# 7. Calculate invoice_value
final_df['invoice_value'] = (
    (final_df['Quantity'] * final_df['Avg_Price']) *
    (1 - final_df['Effective_Discount_pct']) *
    (1 + final_df['GST']) # Assuming GST is applied after discount
) + final_df['Delivery_Charges']

# 8. Assign the new invoice_value column back to the original online_Sales_df
online_Sales_df['invoice_value'] = final_df['invoice_value']
# Also assign the intermediate calculation columns for display
online_Sales_df['Effective_Discount_pct'] = final_df['Effective_Discount_pct']
online_Sales_df['GST'] = final_df['GST']

# Clean up temporary columns from online_Sales_df
online_Sales_df.drop(columns=['Transaction_Month'], inplace=True)

print("invoice_value column added to online_Sales_df. Here's a sample:")
display(online_Sales_df[['Transaction_ID', 'Quantity', 'Avg_Price', 'Coupon_Status', 'Effective_Discount_pct', 'GST', 'Delivery_Charges', 'invoice_value']].head())
```

#**3. Identify periods with the strongest and weakest retention rates. What strategies could be implemented to improve retention during weaker months?**

```python
online_Sales_df['Transaction_Date'] = pd.to_datetime(online_Sales_df['Transaction_Date'])

online_Sales_df['Transaction_Month'] = online_Sales_df['Transaction_Date'].dt.to_period('M')
monthly_active_customers = online_Sales_df.groupby('Transaction_Month')['CustomerID'].nunique().reset_index()
monthly_active_customers.rename(columns={'CustomerID': 'Active_Customers_Count'}, inplace=True)

print("Number of unique active customers per month:")
display(monthly_active_customers.head())
```

```python
customers_by_month = online_Sales_df.groupby('Transaction_Month')['CustomerID'].apply(set).to_dict()

print("Dictionary of unique customers per month (first 2 entries):")
# Display the first few entries of the dictionary for verification
for month, customer_ids in list(customers_by_month.items())[:2]:
    print(f"Month: {month}, Number of unique customers: {len(customer_ids)}")
    print(f"Sample Customer IDs: {list(customer_ids)[:5]}...")
```

```python
retention_data = []
sorted_months = sorted(customers_by_month.keys())

for i in range(1, len(sorted_months)):
    current_month_period = sorted_months[i]
    previous_month_period = sorted_months[i-1]

    current_month_customers = customers_by_month[current_month_period]
    previous_month_customers = customers_by_month[previous_month_period]

    retained_customers = len(previous_month_customers.intersection(current_month_customers))
    previous_month_customers_count = len(previous_month_customers)

    retention_rate = (retained_customers / previous_month_customers_count) * 100 if previous_month_customers_count > 0 else 0

    retention_data.append({
        'Month': current_month_period.strftime('%Y-%m'),
        'Retained_Customers': retained_customers,
        'Previous_Month_Customers_Count': previous_month_customers_count,
        'Retention_Rate': retention_rate
    })

monthly_retention_df = pd.DataFrame(retention_data)

print("Monthly Customer Retention Rates:")
display(monthly_retention_df.head())
```

```python
plt.figure(figsize=(12, 7))
plt.bar(monthly_retention_df['Month'], monthly_retention_df['Retention_Rate'], color='teal')
plt.title('Monthly Customer Retention Rate')
plt.xlabel('Month')
plt.ylabel('Retention Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

highest_retention_month = monthly_retention_df.loc[monthly_retention_df['Retention_Rate'].idxmax()]
lowest_retention_month = monthly_retention_df.loc[monthly_retention_df['Retention_Rate'].idxmin()]

print(f"Month with highest retention rate: {highest_retention_month['Month']} with {highest_retention_month['Retention_Rate']:.2f}% retention")
print(f"Month with lowest retention rate: {lowest_retention_month['Month']} with {lowest_retention_month['Retention_Rate']:.2f}% retention")
```

#**5. Compare the revenue generated by new and existing customers month-over-month. What does this trend suggest about the balance between acquisition and retention efforts?**

```python
import numpy as np

# 1. Extract the year and month from the 'First_Transaction_Date' in the first_transaction_date_df
first_transaction_date_df['Acquisition_Month_Year'] = first_transaction_date_df['First_Transaction_Date'].dt.to_period('M')

# 2. Merge the online_Sales_df with the first_transaction_date_df on 'CustomerID'
online_Sales_df = pd.merge(
    online_Sales_df,
    first_transaction_date_df[['CustomerID', 'Acquisition_Month_Year']],
    on='CustomerID',
    how='left'
)

# Ensure Transaction_Date is in Period('M') format for comparison
online_Sales_df['Transaction_Month_Year'] = online_Sales_df['Transaction_Date'].dt.to_period('M')

# 3. Create a new column named 'Customer_Type' in online_Sales_df.
# 4. For each row in online_Sales_df, if the year and month of the 'Transaction_Date'
#    match the 'Acquisition_Month_Year', classify the customer for that transaction as 'New'.
#    Otherwise, classify them as 'Existing'.
online_Sales_df['Customer_Type'] = np.where(
    online_Sales_df['Transaction_Month_Year'] == online_Sales_df['Acquisition_Month_Year'],
    'New',
    'Existing'
)

# Drop the temporary columns used for merging and comparison
online_Sales_df.drop(columns=['Acquisition_Month_Year', 'Transaction_Month_Year'], inplace=True)

# 5. Display the head of the modified online_Sales_df to verify the new 'Customer_Type' column.
print("Online Sales DataFrame with 'Customer_Type' column:")
display(online_Sales_df.head())
```

#10. Segment customers into groups such as Premium, Gold, Silver, and Standard. What targeted strategies can be developed for each segment to improve retention and revenue? (Use RFM segmentation techniques)

```python
import pandas as pd
import numpy as np

# STEP 1: Set a reference date to calculate Recency
# Use the already defined snapshot_date from the notebook
snapshot_date = online_Sales_df['Transaction_Date'].max()

# STEP 2: Group data by CustomerID and calculate:
# - Recency: days since last purchase
# - Frequency: number of purchases (using Transaction_ID to count unique transactions)
# - Monetary: total value spent (using invoice_value)
alternative_rfm_df = online_Sales_df.groupby('CustomerID').agg({
    'Transaction_Date': lambda x: (snapshot_date - x.max()).days,   # Recency
    'Transaction_ID': 'nunique',                                   # Frequency (count unique transaction IDs)
    'invoice_value': 'sum'                                       # Monetary
}).reset_index()

# Rename columns for clarity
alternative_rfm_df.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

print("Alternative RFM DataFrame (Recency, Frequency, Monetary):")
display(alternative_rfm_df.head())
```

```python
# STEP 3: Create R, F, M segment scores using binning

# Define example thresholds (can be customized or made dynamic)
# Adjusting bins slightly based on previous rfm_df distribution to get similar scoring range
r_bins = [alternative_rfm_df['Recency'].min()-1, 50, 150, 250, alternative_rfm_df['Recency'].max()+1]
f_bins = [alternative_rfm_df['Frequency'].min()-1, 5, 15, 30, alternative_rfm_df['Frequency'].max()+1]
m_bins = [alternative_rfm_df['Monetary'].min()-1, 1000, 5000, 10000, alternative_rfm_df['Monetary'].max()+1]

# Assign segment scores using pd.cut
alternative_rfm_df['R_Score'] = pd.cut(alternative_rfm_df['Recency'], bins=r_bins, labels=False, right=False) + 1
alternative_rfm_df['F_Score'] = pd.cut(alternative_rfm_df['Frequency'], bins=f_bins, labels=False, right=False) + 1
alternative_rfm_df['M_Score'] = pd.cut(alternative_rfm_df['Monetary'], bins=m_bins, labels=False, right=False) + 1

# If any scores result in NaN due to value outside bins, set to a default (e.g., 1)
alternative_rfm_df['R_Score'] = alternative_rfm_df['R_Score'].fillna(1).astype(int)
alternative_rfm_df['F_Score'] = alternative_rfm_df['F_Score'].fillna(1).astype(int)
alternative_rfm_df['M_Score'] = alternative_rfm_df['M_Score'].fillna(1).astype(int)

print("Alternative RFM DataFrame with R, F, and M Scores:")
display(alternative_rfm_df.head())
```

```python
# STEP 4: Combine R, F, M scores to compute an overall segment
alternative_rfm_df['RFM_Score_Combined'] = alternative_rfm_df['R_Score'] + alternative_rfm_df['F_Score'] + alternative_rfm_df['M_Score']

# Create a new customer segment based on RFM_Score_Combined thresholds
def classify_customer(score):
    if score <= 6:
        return 'Standard'
    elif score <= 9:
        return 'Silver'
    elif score <= 11:
        return 'Gold'
    else:
        return 'Premium'
alternative_rfm_df['Customer_Segment'] = alternative_rfm_df['RFM_Score_Combined'].apply(classify_customer)

print("Alternative RFM DataFrame with Combined Score and Customer Segment:")
display(alternative_rfm_df.head())

print("\nDistribution of Customer Segments (Alternative Approach):")
display(alternative_rfm_df['Customer_Segment'].value_counts())
```

#14. Do customers who use coupons have a different average transaction value compared to those who do not? Conduct a statistical test to validate this hypothesis. What implications does this have for the company's discount and coupon strategies?

```python
coupon_used_sales_df = online_Sales_df[online_Sales_df['Coupon_Status'] == 'Used']
coupon_not_used_sales_df = online_Sales_df[online_Sales_df['Coupon_Status'].isin(['Clicked', 'Not Used'])]

print("Coupon Used Sales DataFrame (first 2 rows):")
display(coupon_used_sales_df.head(2))
print("\nCoupon Not Used Sales DataFrame (first 2 rows):")
display(coupon_not_used_sales_df.head(2))
```

## Formulate Hypotheses

### Hypotheses:

*   **Null Hypothesis (H0):** There is no statistically significant difference in the average transaction value between customers who used coupons and customers who did not use coupons.
    *   Mathematically: \( \mu_{\text{used}} = \mu_{\text{not used}} \)

*   **Alternative Hypothesis (H1):** There is a statistically significant difference in the average transaction value between customers who used coupons and customers who did not use coupons.
    *   Mathematically: \( \mu_{\text{used}} \neq \mu_{\text{not used}} \)

```python
from scipy.stats import ttest_ind

# Extract invoice_value for 'coupon used' group
coupon_used_values = coupon_used_sales_df['invoice_value']

# Extract invoice_value for 'coupon not used' group
coupon_not_used_values = coupon_not_used_sales_df['invoice_value']

# Perform independent two-sample t-test
t_statistic, p_value = ttest_ind(coupon_used_values, coupon_not_used_values, equal_var=False) # Assuming unequal variances

print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3e}")
```

### Interpretation of the T-Test Results:

*   **T-statistic: -14.024**
    *   The negative sign indicates that the mean `invoice_value` for the 'coupon used' group is lower than that for the 'coupon not used' group.
    *   The magnitude of the t-statistic (absolute value of 14.024) is large, suggesting a substantial difference between the group means.

*   **P-value: 1.406e-44**
    *   This p-value is extremely small (approaching zero) and is significantly less than any conventional significance level (e.g., 0.05 or 0.01).

### Conclusion:

Given the extremely low p-value (1.406e-44 < 0.05), we **reject the null hypothesis**.

This means there is statistically significant evidence to conclude that there *is a difference* in the average transaction value between customers who use coupons and those who do not.

Specifically, customers who use coupons (`coupon_used_sales_df`) tend to have a **lower average transaction value** compared to customers who do not use coupons (`coupon_not_used_sales_df`). This aligns with the descriptive statistics observed earlier, where the average invoice value for 'Used' coupons was ~$87.93, while 'Clicked' and 'Not Used' were ~$108.66 and ~$110.86 respectively.

### Implications for the Company's Discount and Coupon Strategies:

1.  **Reinforces Segmented Approach:** The statistical difference confirms that coupon users are a distinct segment with different purchasing behaviors. Strategies should acknowledge this. Customers using coupons are likely more price-sensitive or are encouraged to purchase lower-priced items when a discount is available.

The company needs to carefully assess if the increased volume from coupon usage sufficiently offsets the reduction in average revenue per transaction to maintain or improve overall profitability.     *   **Targeted Promotions:** Instead of broad discounts, consider targeted coupons to specific customer segments (e.g., to reactivate lapsed customers, to encourage a second purchase from a first-time buyer in a low-CLTV cohort) rather than blanket promotions that might dilute revenue from less price-sensitive customers.
